# Period 9 Post Survey

```python
import pandas as pd
df1 = pd.read_csv('Class Post -Survey (Responses) - Form Responses 1.csv')
```

```python
df1.columns
```

## Concat both data set to compare pre: 6a and post: 6a
- need to change the name of the columns to make sure to keep it different

```python
# rename 6a in df to pre
df.rename(columns = {'6a. Based on your answers in this survey, rate your level of responsibility as a student in THIS science class with Mrs. Rojas. (1-very low, 2-somewhat low, 3-average, 4-somewhat high, 5-very high)': 'pre'},
                    inplace=True)
```

```python
# rename 6a in df1 to post
df1.rename(columns = {'6a. Based on your answers in this survey, rate your level of responsibility as a student in THIS science class with Mrs. Rojas. (1-very low, 2-somewhat low, 3-average, 4-somewhat high, 5-very high)' : 'post'},
                    inplace=True)
```

```python
compare = pd.concat([df, df1], axis = 1)
```

```python
compare_df = compare[['pre', 'post']]
compare_df.info()
```

```python
# fill the missing value with the average - 4
compare_df['post'].fillna(value=compare_df['pre'].median(), inplace=True)
```

```python
# drop the .0
compare_df['post'] = compare_df.post.astype(int)
```

## Plot the comparison
- value counts
- bar graph

```python
print('pre')
display(compare_df.pre.value_counts())
print('post')
display(compare_df.post.value_counts())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


fig1 = plt.figure(figsize=(15,6))
# value counts for bars 1, 2, 3, 4, 5
barWidth = 0.25
bars1 = [0, 2, 9, 10, 7]
bars2 = [0, 2, 7, 13, 6]


r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]



plt.bar(r1, bars1, color='darkmagenta', width=barWidth, edgecolor='white', label='Pre Survey')
plt.bar(r2, bars2, color='violet', width=barWidth, edgecolor='white', label='Post Survey')


# Add xticks on the middle of the group bars
# plt.title('Comparing Student Preception of Responsibilities in Science Class - 45 Day Difference', fontweight='bold', fontsize=16)
plt.xlabel('Responses', fontweight='bold', fontsize=12)
plt.xticks([r + barWidth for r in range(len(bars1))], ['1', '2', '3', '4', '5'], fontsize=12)
plt.ylabel('Number of Students', fontweight='bold', fontsize=12)
 
# Create legend & Show graphic
plt.legend()
plt.show()
fig1.savefig('plot1.png')
```
